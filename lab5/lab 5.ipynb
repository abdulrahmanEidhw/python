## data visualaisation

```python
for ojha, feature in enumerate(list(df_with_columns.columns)[:-1]):
    fg = sns.FacetGrid(df_with_columns, hue='x4', height=5)
    fg.map(sns.distplot, feature).add_legend()
    plt.show()
```

```python
sns.boxplot(x='x4', y='x', data=df_with_columns)
plt.show()
```

```python
sns.boxplot(x='x4',y='x',data=df_with_columns)
plt.show()
```

```python
data=df_with_columns.plot(kind='sctter',,x='sepal length(cm)',y='sepal width (cm)',c='g')
plt.show
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df_with_columns, hue="Variety",height=10) \
    .map(plt.scatter, "x4","x") \
    .add_legend();
plt.show();
```

```python
sns.set_style("whitegrid");
sns.pairplot(df_with_columns,vars = iris.feature_names, hue = 'Variety', diag_kind = 'kde',
              plot_kws={'alpha':0.6, 's':80, 'edgecolor': 'k'}, size =2)
plt.show()
```
